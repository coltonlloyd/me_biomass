```python
import pandas as pd
import numpy as np
import cobra
from collections import defaultdict
```

```python
df = pd.read_excel('Supplementary Data File 11 - Essentiality.xls', sheet_name=" Table 2").set_index('Gene')
```

```python
measure_dict = defaultdict(list)
compute_dict = defaultdict(list)
for i, temp_df in df.groupby('Condition'):
    for gene in temp_df.index:
        meas_val = temp_df.loc[gene, 'Measured_growth']
        meas_e = 0 if meas_val < .5 else 1
        comp_val = temp_df.loc[gene, 'Computational_growth']
        comp_e = 0 if comp_val < .1 else 1
        
        measure_dict[gene].append(meas_e)
        compute_dict[gene].append(comp_e)
```

```python
ijo = cobra.io.load_json_model('/home/sbrg-cjlloyd/Desktop/ecoli_M_models/iJO1366.json')
```

```python
for gene in compute_dict:
    values = np.array(compute_dict[gene])
    meas_values = np.array(measure_dict[gene])
    if values.sum() == 0 or values.sum() == len(values):
        if meas_values.sum() != 0 and meas_values.sum() < len(values):
            print(gene, ijo.genes.get_by_id(gene).name)
```
